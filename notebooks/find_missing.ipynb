```python
import difflib
import os
from bs4 import BeautifulSoup
import requests
from collections import Counter
```

```python
with open("./outputs/html.txt") as f:
    soup = BeautifulSoup(f.read(), "html.parser")

playlist = soup.find_all(
    "ytmusic-playlist-shelf-renderer",
    class_="style-scope ytmusic-section-list-renderer fullbleed",
)
songs = playlist[0].find_all(
    "ytmusic-responsive-list-item-renderer",
    class_="style-scope ytmusic-playlist-shelf-renderer",
)
titles = []
for song in songs:
    title = (
        song.find(
            "div",
            class_="flex-columns style-scope ytmusic-responsive-list-item-renderer",
        )
        .find("div", "title-column style-scope ytmusic-responsive-list-item-renderer")
        .find("yt-formatted-string")["title"]
    )
    titles.append(title)
len(titles), titles
```

```python
count = Counter(titles)
count
```

```python
def find_similar_files(names, directory, cutoff=0.6):
    files_in_directory = os.listdir(directory)
    files_in_directory = [file.lower().rsplit(".", 1)[0] for file in files_in_directory]
    similar_files = {}

    print(f"{len(files_in_directory)} files.")
    # print(Counter(files_in_directory))
    print(f"{len(names)} names.")
    for name in names:
        matches = difflib.get_close_matches(
            name.lower(), files_in_directory, cutoff=cutoff
        )
        if matches:
            similar_files[name] = matches
        else:
            similar_files[name] = []

    return similar_files
```

```python
directory_path = "../../files/Classical//"

similar_files = find_similar_files(titles, directory_path, 0.8)

# Print the results
not_found = []
found = []

for name, matches in similar_files.items():
    if matches:
        found.append(name + ": -----" + str(matches))
        # print(f"Files similar to '{name}': {matches}")
    else:
        not_found.append(name)
        print(f"Not found: {name}")


print(f"Found: {len(found)}")
print(f"Not found: {len(not_found)}")
```
